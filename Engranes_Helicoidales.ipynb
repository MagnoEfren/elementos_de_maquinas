```python
import numpy as np
import math 

# VARIABLES 
angulo_psi = np.radians(20)  # Angulo de la helice   
angulo_fi_n = np.radians(20)   # angulo de presión normal
angulo_fi_t = np.arctan(np.tan(angulo_fi_n)/np.cos(angulo_psi)) # angulo de presión trasversal
P = 8 # Asignamos un paso diametral estandar  de 8 dientes/pulg para las 2 etapas (4 engranes).
k = 1 # Dientes de profundidad completa

def calcular_numero_dientes(n_5, n_2):
    m = round((n_5/n_2),2)
    # N_P  es el minimo numero de dientes del piñon para evitar interferecia con el engrane acoplado, (Ec: 13.11 Shigley- Pag. 660)
    N_P =  math.ceil(((2*k)/((1+2*m)*np.sin(angulo_fi_n)**2))*(m + np.sqrt(m**2 + (1+2*m)*np.sin(angulo_fi_n)**2)))
    N_5 = N_P
    """Para minimizar el tamaño de la caja, se mantienen las partes tan equitativamente 
    divididas entre las etapas como sea posible. En los casos donde el valor del 
    tren total solamente necesita ser aproximado, cada etapa puede ser idéntica. 
    Por ejemplo, en un tren de engranes compuesto de dos etapas, se asigna la raíz (dedendum) 
    cuadrada del valor del tren total a cada etapa."""
    N_4 = int(N_5*np.sqrt(m))
    N_3 = math.ceil((N_5 + N_4)/(1+ np.sqrt(m)))  # N_2 = np.sqrt(m)*N_3
    N_2 = int(N_3*np.sqrt(m))
    return N_2, N_3, N_4, N_5

while True: 
    try: 
        n_5 =  int(input('Ingrese la velocidad de entrada en RPM: '))  # (Sentido horario) 
        n_2 = int(input('Ingrese la velocidad de salida en RPM: '))  # velocidad de salida en rpm en el punto 2
        HP  = int(input('Ingrese la potencia en HP: ')) # Potencia en HP  podemos considerar un rango en este caso solo se considero que debe ser positiva 
        if 0 < HP and n_5 > 150  and 150 <= n_2 and n_2 <= 300:          
            N_x = calcular_numero_dientes(n_5, n_2)
            N_2 = N_x[0]
            N_3 = N_x[1]
            N_4 = N_x[2]
            N_5 = N_x[3]
            print('Numero de dientes del engrane 2: ', N_2)
            print('Numero de dientes del engrane 3: ', N_3)
            print('Numero de dientes del engrane 4: ', N_4)
            print('Numero de dientes del engrane 5: ', N_5)
            print('\t')
            # ENGRANE 5
            d_5 = N_5/P #Diametro del engrane
            W_t_5 = round(63000*(HP/(n_5*(d_5/2))), 4)  # tangencial 
            W_r_5 = round(W_t_5*np.tan(angulo_fi_t), 4)  # radial 
            W_a_5 = round(W_t_5*np.tan(angulo_psi),4)  #axial 
            W_5 = round((W_t_5)/(np.cos(angulo_fi_n)*np.cos(angulo_psi)), 4) # Fuerza total
            T_5 = round(W_t_5*(d_5/2),4) # Par torcional de entrada

            print('Fuerza tangencial en el engrane 5: ', W_t_5, 'lbf')
            print('Fuerza radial en el engrane 5: ', W_r_5, 'lbf')
            print('Fuerza  transmitida en el engrane 5: ', W_a_5, 'lbf')
            print('Fuerza total en el engrane 5: ', W_5, 'lbf')
            print('Par torcion de  entrada: ', T_5, 'lbf*pulg')
            print('\t')
            # ENGRANE 4    
            d_4 = N_4/P #Diametro del engrane
            n_4 = (n_5*d_5)/d_4
            W_t_4 = W_t_5  # Es igual por accion y reaccion entre los engranes 5 y 4 
            W_r_4 = round(W_t_4*np.tan(angulo_fi_t),4)  # radial 
            W_a_4 = round(W_t_4*np.tan(angulo_psi),4)  #axial 
            W_4 = round((W_t_4)/(np.cos(angulo_fi_n)*np.cos(angulo_psi)),4) # Fuerza total
            T_4 = round(T_5*(d_4/d_5), 4) # Par torcional del engrane 4

            print('Fuerza tangencial en el engrane 4: ', W_t_4, 'lbf')
            print('Fuerza radial en el engrane 4: ', W_r_4, 'lbf')
            print('Fuerza  transmitida en el engrane 4: ', W_a_4, 'lbf')
            print('Fuerza total en el engrane 4: ', W_4, 'lbf')
            print('Par torcion en el engrane 4: ', T_4, 'lbf*pulg')
            print('\t')

            # ENGRANE 3
            d_3 = N_3/P #Diametro del engrane
            n_3 = n_4 # Estan sobre el mismo eje por ello giran a la misma velocidad
            T_3 = T_4 # Par torcional del engrane 3
            W_t_3 = round( (2*(T_3/d_3)),4)   # tangencial 
            W_r_3 = round(W_t_3*np.tan(angulo_fi_t),4)  # radial 
            W_a_3 = round(W_t_3*np.tan(angulo_psi),4)   #axial 
            W_3 = round((W_t_3)/(np.cos(angulo_fi_n)*np.cos(angulo_psi)),4)  # Fuerza total

            print('Fuerza tangencial en el engrane 3: ', W_t_3, 'lbf')
            print('Fuerza radial en el engrane 3: ', W_r_3, 'lbf')
            print('Fuerza  transmitida en el engrane 3: ', W_a_3, 'lbf')
            print('Fuerza total en el engrane 3: ', W_3, 'lbf')
            print('Par torcion en el engrane 3: ', T_3, 'lbf*pulg')
            print('\t')

            # ENGRANE 2
            d_2 = N_2/P #Diametro del engrane
            W_t_2  = W_t_3  # tangencial 
            W_r_2 = round(W_t_2*np.tan(angulo_fi_t),4)   # radial 
            W_a_2 = round(W_t_2*np.tan(angulo_psi),4)   #axial 
            W_2 = round((W_t_2)/(np.cos(angulo_fi_n)*np.cos(angulo_psi)),4)  # Fuerza total
            T_2 = round((T_3)*(d_2/d_3),4) # Par torcional de salida

            print('Fuerza tangencial en el engrane 2: ', W_t_2, 'lbf')
            print('Fuerza radial en el engrane 2: ', W_r_2, 'lbf')
            print('Fuerza  transmitida en el engrane 2: ', W_a_2, 'lbf')
            print('Fuerza total en el engrane 2: ', W_2, 'lbf')
            print('Par torcion de salida: ', T_2, 'lbf*pulg')
            print('\t')

            # Verificamos que la velocidad de linea del conjunto 1 (engranes 4 y 5) es el mismo    
            V_L5 = round(((np.pi*d_5*n_5)/12), 4)  #solo para verificar
            V_L4 = round((np.pi*d_4*n_4)/12, 4)   #solo para verificar 
            #Encontramos que V_L4=V_L5
            V_Lconjunto1=V_L4 #Velocidad le linea del acople engrane 5-4
            print('Velocidad lineal en el conjunto 1: ', V_Lconjunto1, 'pies/min' )
            # Verificamos que la velocidad de linea del conjunto 2 (engranes 2 y 3) es el mismo
            V_L3 = round((np.pi*d_3*n_3)/12, 4) #solo para verificar 
            V_L2 = round((np.pi*d_2*n_2)/12, 4) #solo para verificar 
            #Encontramos que V_L2 = V_L3
            V_Lconjunto2=V_L2 #Velocidad le linea del acople engrane 3-2
            print('Velocidad lineal en el conjunto 2: ', V_Lconjunto2, 'pies/min' )
            print('\t')
            break
        else:
            print("Error al ingresar los parametros de entrada: \n -Velocidad de entrada considere  valores mayores a 150  \n -Velocidad de salida considere  valores entre 150 y 300 \n -La potencia en HP debe ser mayor a 0")
            x = input('Ingrese 0 par salir o 1 para continuar: ')   
            if x =='0':
                break
            else:
                continue
    except ValueError:
        print("Error: Debes ingresar un valor numerico correcto")
       
```
